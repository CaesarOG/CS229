<h1>What is Machine Learning?</h1>

It is the science of getting computers to learn, without being explicitly programmed. Arthur Samuel 1959 (made checkers program better than him).

A computer program is said to learn from experience E with respect to some task T and performance measure P, if its performance on T, as measured by P, improves with experience E. Tom Mitchell 1998. 

In checkers, E is 10s of 1000s of games, T is playing checkers, P is probability of winning next game against new opponent. If marking email as spam or not, and learnds off of you to better filter, here T is classifying emails as spam or not. 

Q: suppose email program watches you mark as spam or not, and learns to better filter it. What is task T? A. marking as spam or not spam, B. watching you mark them as such, C. the number/fraction correctly classified as spam or not, or D. none of the above it's not ML? 

A: choice A.

Time to learn main ML terms, sense of different algs and where each appropriate. Main two types are supervised learning (teach from experience) and unsupervised learning (learn to find some structure on its own) . Other kinds are reinforcement learning, or recommender systems. Equally important to giving tools is practical advice how to apply, best practices.  Not just "here's a hammer, now ur a carpenter".

<h1>Supervised Learning</h1>

```octave
I=im2double(imread('./files/housing.png'));
imshow(I)
```

This comes from a visualization of Portland housing market several years ago. You have a friend that owns house 750 sqft, wants to know how much to sell for. Learning algorithm could draw line/curve through data to predict where other points could end up. You can see the prediction from a line in pink, and a second order polynomial (quadratic) curve in blue, with their respective predictions. 

We'll later talk about how to decide do u want straight or quadratic through the points. The right answers for every example are given in supervised learning. Then task T is to produce more right answers for new examples (new x values in sqft).  

Supervised learning problems are categorized into "regression" and "classification" problems. In a regression problem, we are trying to predict results within a continuous output, meaning that we are trying to map input variables to some continuous function. In a classification problem, we are instead trying to predict results in a discrete output. In other words, we are trying to map input variables into discrete categories. Above example's clearly regression. 

```octave
I=im2double(imread('./files/diagnose.png'));
imshow(I)
```

You want to predict whether malignant or benign given just tumor size. This is classification, since unlike the continuous value output/"answers", here we're predicting out of a range of discrete outputs that are possible answers. The basic example shows tumor size vs. malignant? (0 for no a.k.a benign, 1 for yes). Given the one input, output probability that it is 0, probability that it's 1. 

Could have more classes as shown, ie. 3 types of cancers, so 0 still benign/no, with 1 now representing type 1, 2 is type 2, 3 is type 3. In classification there is another way to plot, you can get rid of the axis for the output, and since only one input (tumor size), it's just a one axis graph, mapping over from the first graph with 'O' for benign 'X' for malignant. 

```octave
I=im2double(imread('./files/diagnose2.png'));
imshow(I)
```

Lets use this classification-specific graph type with two inputs, age and tumor size. The pink shows someone who has that size and age. The learning alg might fit a straight line to separate them as shown. Output would show higher probability for 0 (benign), so more likely benign. 

Many ML problems will have still more features. In this particular real world version, clump thickness, uniformity of cell size and uniformity of cell shape are also features for BC diagnose, among others. 

One of the most interesting learning algs we'll see in this class, can deal with not just 1, 2 or 5 features, but infinite number of features. How do you deal with that, can't install infinite things on computer will run out of memory. Support Vector Machine allows that, uses kernel trick to map to lower dimension space less features so it is actually computationally tractable to calc the boundary.

<h1>Unsupervised Learning</h1>

```octave
I=im2double(imread('./files/unsupervised.png'));
imshow(I)
```

In unsupervised learning given data with no labels or all one label. Idea is, here is dataset, can you find some structure in it. Clustering algorithm is one, example above it finds that data lives in two clusters. Used in many places, one is Google News. Looks at 10s or 100s of thousands of articles a day and groups into cohesive stories. The links under each headline are such that all on same topic displayed together, Google News automatically clustered them together. 

```octave
I=im2double(imread('./files/unsupervised2.png'));
imshow(I)
```

In genomics you have dna microarray data like above, 2d data sets of genes and individuals measuring the degree to which do or don't have a gene (how much one is expressed), colors show the degree. You can run clustering, it would group into diff types of people Once clusters found can decide if meaningful/helpful and give labels. Saying don't know what's in it, don't know who's in what type hell don't know what the types are. Not giving the "right answers" ie. group 1 for the examples, so unsupervised.

Another example is large computer centers, if you can find what machines tend to work together, and if you put those more together, center will be more efficient. Also social network analysis, like given knowledge about frequent friends/emails identify groups that all know each other. Market segmentation, given huge customer dataset, automatically group for more efficient marketing (dont know what the segments are and who in what one). Even astronomical data, clustering help give surprisingly useful theories on how galaxies formed for example. All these are examples where clustering algorithms can help.

Another type of unsupervised algorithm, indep component analysis, can be explained with the cocktail party problem. All these overlapping voices, almost hard to hear person in front of you. Say just two people talking together same time, two microphones at two diff distances from the speakers. Each mic picks up diff combo, say speaker 1 louder in mic 1 and speaker 2 in mic 2. Or maybe 1 person and 1 radio playing music. Take both recordings and give it to the algorithm to find structure, it will find that the two same sources are being summed together in different intensities in both mics, and moreover it will separate out the two sources and give those two as the output. 

You might think it would take a lot to do this, for pulling in and parsing audio and whatever else for the learning alg that actually separates it maybe need C++ or Java, but it turns out it's one line of code in Octave. 

\[W,s,v\] = svd((repmat(sum(x.\*x,1),size(x,1),1).\*x)\*x'); 

Now it did take researchers long time to come up with it, it's not easy, but point is in right environment many programs can be very short. Many learning algs just few lines. If you look at the line of code, functions like svd (singular value decomposition, a lin alg routine) and repmat, are just built into octave. So can use C++ or Java or Python, but it would be many lines of code linking complex c++ or java libraries, cumbersome to prototype in. Having taught ML for a decade+, Ng has seen that you learn much faster if you use Octave as learning and prototyping tool. So in Silicon Valley large companies often prototype in octave and then after it works port it to C++ or Java or even Tensorflow(which uses NumPy's C code). Your developer time is one of the most valuable resources, Ng has seen lot of people save lot of time by prototyping and learning in Octave. 

Q: which problems would use unsupervised learning algorithms? A. Given email labeled spam or not learn filter for email, B. given bunch of articles group into sets on same story, C. Given customer database discover market segments to group customers, D. given patients diagnosed diabetes or not learn to say if have it. Answer B and C, others are clearly supervised have output answers for all inputs and want to predict for new input. 

Summary: 

Unsupervised learning allows us to approach problems with little or no idea what our results should look like. We can derive structure from data where we don't necessarily know the effect of the variables. We can derive this structure by clustering the data based on relationships among the variables in the data.
With unsupervised learning there is no feedback based on the prediction results.

Clustering example: Take a collection of 1,000,000 different genes, and find a way to automatically group these genes into groups that are somehow similar or related by different variables, such as lifespan, location, roles, and so on.
Non-clustering example: The "Cocktail Party Algorithm", allows you to find structure in a chaotic environment. (i.e. identifying individual voices and music from a mesh of sounds at a cocktail party).


<h1>Model Representation</h1> 

Time to see what supervised learning looks like. We'll start learning how to make a model by doing the most simple one, linear regression with one input variable/feature. Back to housing prices in Portland in [Supervised Learning](#Supervised-Learning) section. The straight line of best fit was linear regression. Remember it's regression since real-valued continuous and not discrete (which is classification). 

So have a training set, with `m` training examples, `x's` are input variable/features, and `y's` are output/target variable. $(x,y)$ is one training example, $(x^{(i)}, y^{(i)})$ is the $i^{th}$ example. That superscript is not power, just sample index, note it's not 0-based like in programming, but natural numbers 1 through m. A capital $X$ is a vector of all `x's`, and similar for capital $Y$. So, use X to denote the space of input values, and Y to denote the space of output values. In this example, X = Y = ℝ.
See below picture.

```octave
I=im2double(imread('./files/modelrep.png'));
imshow(I)
```

Feed the training set to a learning algorithm, whose job it is to learn a function usually called $h$ for hypothesis, which takes size of house and outputs estimated price. So $h$ maps from `x's` to `y's` such that it's a good predictor of y. Calling it hypothesis is just a standard since start of ML, don't worry about the name. 

$h_{\theta}(x) = \theta_{0} + \theta_{1}x$ is the formal definition, shorthand for $h_{\theta}(x)$ is $h(x)$. Doing linear function first. Sometimes need to fit nonlinear functions as well, but linear case is building block. 
Since only input variable/feature, x, it is linear reg with one var, or univariate lin reg.
